```python
from skimage import data, color
import matplotlib.pyplot as plt
import numpy as np
from skimage.transform import resize, rotate
from skimage.filters import gaussian, threshold_otsu
from skimage.feature import canny
from google.colab import files

# Upload image file manually in Google Colab
uploaded = files.upload()

# Load the uploaded image
madrid_image = plt.imread('demon.png')

# Convert RGBA to RGB if necessary
if madrid_image.shape[-1] == 4:
    madrid_image = madrid_image[:, :, :3]

# Convert to grayscale
grayscale = color.rgb2gray(madrid_image)
rgb = color.gray2rgb(grayscale)

# Function to display images
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Extract RGB channels
red = madrid_image[:, :, 0]
green = madrid_image[:, :, 1]
blue = madrid_image[:, :, 2]

# Image properties
print(madrid_image.shape)  # (height, width, channels)
print(madrid_image.size)   # Total number of pixels

# Image transformations
resized_image = resize(madrid_image, (100, 100))
rotated_image = rotate(madrid_image, 45)
flipped_image = np.fliplr(madrid_image)

# Image filtering
blurred_image = gaussian(madrid_image, sigma=1)

# Edge detection
edges = canny(grayscale, sigma=1)

# Thresholding
thresh = threshold_otsu(grayscale)
binary_image = grayscale > thresh
```

```python
# Show original image
show_image(madrid_image, "Original Image")

# Show grayscale image
show_image(grayscale, "Grayscale Image")

# Show reconstructed RGB image from grayscale
show_image(rgb, "Reconstructed RGB from Grayscale")

# Show individual color channels
show_image(red, "Red Channel", cmap_type='Reds')
show_image(green, "Green Channel", cmap_type='Greens')
show_image(blue, "Blue Channel", cmap_type='Blues')

# Show resized image
show_image(resized_image, "Resized Image")

# Show rotated image
show_image(rotated_image, "Rotated Image (45 degrees)")

# Show flipped image
show_image(flipped_image, "Flipped Image (Horizontally)")

# Show blurred image
show_image(blurred_image, "Blurred Image (Gaussian)")

# Show edge-detected image
show_image(edges, "Edge Detection (Canny)")

# Show thresholded binary image
show_image(binary_image, "Thresholded Binary Image")
```
